## Handling Missing Values

```python
#Replace the missing values in Embarked column by 'S'
df.head()

count=df['Embarked'].isnull().sum()
count

df['Embarked'].fillna('S',inplace=True)
df.head()
```

```python
#Impute the missing values in Fare column with median value
count=df['Fare'].isnull().sum()
print(count)

df['Fare'].fillna(df['Fare'].median(),inplace=True)
```

```python
import numpy as np
#Impute the Age column. It has too many missing values but it may be important. 
#We fill the missing values with numbers randomly drawn from distribution
#that has same mean and standard deviation as non missing Age data
c=df['Age'].isnull().sum()
print(c)

# Display missing values count
print("Missing values in Age column before imputation:",c)

# Compute mean and standard deviation of non-missing 'Age' values
mean_age = df['Age'].mean()
std_age = df['Age'].std()

# Generate random values from normal distribution with same mean & std dev
num_missing = c
random_ages = np.random.normal(mean_age, std_age, num_missing)

# Fill missing values with these random numbers
df.loc[df['Age'].isnull(), 'Age'] = random_ages
df
```
